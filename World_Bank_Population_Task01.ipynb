## Data visualization

```python
# Data visualization
def add_value_labels(ax,spacing=10):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3

        va='bottom' if y>=0 else 'top'
        label= '()'.format(y)
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

# Bar plot for the top ten countries of total population from 2000 to 2022


```python
plt.figure(figsize=(15, 8))
sns.barplot(x='2000', y="Country Name", data=top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total Population (2000-2022)")
plt.xlabel("Total Population")
plt.ylabel("Country")
plt.show()
```
